```python
iris = load_iris()
X, y = pd.DataFrame(iris.data), iris.target
```

## **Param1:n_component**

```python
pca = PCA(n_components=2) # 实例化
pca = pca.fit(X)
X_dr = pca.transform(X)
```

```python
X_dr.shape
```

```python
X_dr
```

```python
np.unique(y)
```

```python
X_dr[y == 0, 0] # 标签为0的第一个特征
```

```python
plt.figure()
plt.scatter(X_dr[y == 0, 0], X_dr[y==0, 1], c='red', label=iris.target_names[0])
plt.scatter(X_dr[y == 1, 0], X_dr[y==1, 1], c='blue', label=iris.target_names[1])
plt.scatter(X_dr[y == 2, 0], X_dr[y==2, 1], c='orange', label=iris.target_names[2])
plt.legend()
plt.title('PCA of Iris Dataset')
plt.show()
```

```python
# 新特征向量所带信息量大小（可解释性方差大小）
pca.explained_variance_
```

```python
# 查看降维后每个新特征向量所占信息量占原始数据总信息量的百分比
# 可解释性方差贡献率
pca.explained_variance_ratio_ # 可以看出大部分信息被有效的集中在了第一个特征上
```

```python
pca.explained_variance_ratio_.sum()
# 损失0.03的信息量
```

## 建模提取新特征矩阵

```python
PCA_ = PCA(150).fit(X) # 只接受<=2dim
```

```python
V = PCA_.components_
```

```python
V.shape #新特征空间
```

```python
# 在映射之前选区两哪些特征:五官&光线
fig, axes = plt.subplots(3,8
                        , figsize=(8,4)
                        , subplot_kw={"xticks":[],"yticks":[]} # 不显示坐标轴
                        )
# axes[0][0].imshow(faces.images[0,:,:])
for i, ax in enumerate(axes.flat):
    ax.imshow(V[i,:].reshape(62,47), cmap='gray')
```

```python
x_dr = PCA_.transform(X)
x_dr.shape
```

```python
PCA_.explained_variance_ratio_.shape
```

### 重要接口inverse_transform
过程可逆，即PCA过程可逆

```python
x_inverse = PCA_.inverse_transform(x_dr)
```

```python
x_inverse.shape
```

```python
fig, axes = plt.subplots(2,10
                        , figsize=(10,2.5)
                        , subplot_kw={"xticks":[],"yticks":[]} # 不显示坐标轴
                        )
# axes[0][0].imshow(faces.images[0,:,:])
for i in range(10):
    axes[0, i].imshow(faces.images[i,:,:], cmap='binary_r')
    axes[1, i].imshow(x_inverse[i,:].reshape(62,47), cmap='binary_r') # x_inverse只返回了150个特征 然后升维的图像
```
