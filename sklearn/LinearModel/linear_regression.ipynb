## 分享离散化：帮助线性回归处理非线性数据

```python
rnd = np.random.RandomState(42)
X = rnd.uniform(-3, 3, size=100) # random.uniform, 从输入的任意两个整数中取出size个随机树

y = np.sin(X) + rnd.normal(size=len(X))/3

plt.scatter(X, y, marker='o',c='k',s=20)
plt.show()
```

```python
X.shape
```

```python
X = X.reshape(-1,1)
```

```python
# 使用原始数据建模
LinearR_ = LinearR().fit(X, y)
TreeR = DTR(random_state=0).fit(X, y)
```

```python
fig, ax = plt.subplots(1)
# test data
line = np.linspace(-3, 3, 1000, endpoint=False).reshape(-1, 1)

ax.plot(line, LinearR_.predict(line), linewidth=2, color = 'green', label='linear regression')
ax.plot(line, TreeR.predict(line), linewidth=2, color = 'orange', label='Decision Tree regression')
# 绘制元数据
ax.plot(X ,y, 'o', c='k')

# 其他图形选项
ax.legend(loc='best')
ax.set_ylabel('Regression Output')
ax.set_xlabel('Input feature')
plt.tight_layout()
plt.show()
```

```python
# 分箱改善
from sklearn.preprocessing import KBinsDiscretizer

# 将数据分箱
enc = KBinsDiscretizer(n_bins=10, # 分几类
                       encode='onehot') # 分类后的变量如何处理；为ordinal时有10类则1列显示0-9
x_binned = enc.fit_transform(X)
# encode='onehot':使用哑变量方式做离散化
# 之后返回一个稀疏矩阵，每一列是特征中的一个类别
# 对每一个样本而言，它包含的分类（箱子）中它表示为1，其余分类中表示为0
```

```python
X.shape
```

```python
x_binned
```

```python
pd.DataFrame(x_binned.toarray()).head()
```

```python
LinearR_2 = LinearR().fit(x_binned, y)
```

```python
line_binned = enc.transform(line)
```

```python
ypreds = LinearR_2.predict(line_binned)
```

```python
ypreds.shape
```

```python
TreeR_ = DTR(random_state=0).fit(x_binned, y)
```

```python
fig, (ax1, ax2)= plt.subplots(ncols=2
                            , sharey=True
                            , figsize=(10, 4))

ax1.plot(line, LinearR_.predict(line), linewidth=2, color = 'green', label='linear regression')
ax1.plot(line, TreeR.predict(line), linewidth=2, color = 'orange', label='Decision Tree regression')
ax1.plot(X, y, 'o', c='k')
ax1.legend(loc='best')
ax1.set_ylabel('Regression Output')
ax1.set_xlabel('Input feature')
ax1.set_title('Result before discretization')

ax2.plot(line, LinearR_2.predict(line_binned), linewidth=2, color='green', linestyle='-', label='line regression')
ax2.plot(line, TreeR_.predict(line_binned), linewidth=2, color='orange', linestyle=':', label='decision tree')
# 绘制和箱宽一致的竖线
ax2.vlines(enc.bin_edges_[0]# x axis
         , *plt.gca().get_ylim() # y轴上下限
         , linewidth=1
         , alpha=0.2)

# 将原数据分布放置再图上
ax2.plot(X, y, 'o', c='k')
ax2.legend(loc='best')
# ax2.set_ylabel('Regression Output')
ax2.set_xlabel('Input feature')
ax2.set_title('Result after discretization')
plt.tight_layout()
plt.show()
```
