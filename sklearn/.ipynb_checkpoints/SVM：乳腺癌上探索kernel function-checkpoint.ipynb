## 乳腺癌数据集上探索核函数
实际中往往不是事先知道数据分布

```python
from sklearn.datasets import load_breast_cancer
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np
from time import time  # 记时
import datetime  # 时间戳转换为真实时间
```

```python
data = load_breast_cancer()
X = data.data
y = data.target
```

```python
X.shape
```

```python
np.unique(y)
```

```python
# 降维后可视化
from sklearn.decomposition import PCA
X_dr = PCA(2).fit_transform(X)
X_dr.shape
```

```python
plt.scatter(X_dr[:,0],X_dr[:,1],c=y,cmap='rainbow')
plt.show()
```

样本不均衡问题，线性可分但紫色会部分分错

```python
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X, y, test_size=0.3, random_state=20)
Kernel = ['linear', 'poly', 'sigmoid', 'rbf']
```

```python

for kernel in Kernel:
    time0 = time()
    clf = SVC(kernel=kernel
             ,gamma='auto'
              , degree=1 # 多项式核函数的最大次数，默认3次
              , cache_size=5000  # 允许使用最大的内存 MB
             ).fit(Xtrain,Ytrain)
    print('the accuracy under kernel %s is %f'% (kernel, clf.score(Xtest, Ytest)))
    print(datetime.datetime.fromtimestamp(time()-time0).strftime("%M:%S:%f"))  # 几分
```

- 卡在了第二个循环，poly在维度上计算3次过于复杂， python内核卡死
- 非线性核函数表现并不好，但是linear的计算时间长，用非线性的核函数达到线性的效果->poly with degree=1

为什么rbf的效果差，按理说在线性和非线性数据上的效果都应该好->从数据出发，量纲统一？数据稳定？

```python
import pandas as pd
data = pd.DataFrame(X)
data.describe([0.01,0.05,0.1,0.25,0.5,0.75,0.9,0.99]).T
# 量纲不统一： 均值和方差
# 偏态问题 ：最大值和0.01/0.09比较->正太&偏态？
```

```python
# 标准化解决上述问题
from sklearn.preprocessing import StandardScaler
X = StandardScaler().fit_transform(X)
```

```python
data = pd.DataFrame(X)
data.describe([0.01,0.05,0.1,0.25,0.5,0.75,0.9,0.99]).T
```

```python
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X, y, test_size=0.3, random_state=20)
Kernel = ['linear', 'poly', 'sigmoid', 'rbf']
for kernel in Kernel:
    time0 = time()
    clf = SVC(kernel=kernel
             ,gamma='auto'
              , degree=1 # 多项式核函数的最大次数，默认3次
              , cache_size=5000  # 允许使用最大的内存 MB
             ).fit(Xtrain,Ytrain)
    print('the accuracy under kernel %s is %f'% (kernel, clf.score(Xtest, Ytest)))
    print(datetime.datetime.fromtimestamp(time()-time0).strftime("%M:%S:%f"))  # 几分
```

因为支持向量机是对距离进行计算，在高维空间计算，因此可以得出**采用SVM前先对数据进行标准化，无量纲化**，conculsion：
- 线性核，尤其是多项式核函数在高次项计算时非常缓慢
- rbf和poly都不擅长处理量纲不统一的数据

线性核除了kernel之外无参数调整，rbf还有其余参数(gamma, degree, coef0)进行调整，但是难以直观看出影响，采用网格搜索和学习曲线查找最佳的参数组合

```python
# 学习曲线rbf
score = []
gamma_range = np.logspace(-5,0,50)
for i in gamma_range:
    clf = SVC(kernel='rbf',gamma=i,cache_size=5000).fit(Xtrain,Ytrain)
    score.append(clf.score(Xtest,Ytest))
print(gamma_range[score.index(max(score))],max(score))
plt.plot(gamma_range, score)
plt.show()
```

```python
# 网格搜数poly
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.model_selection import GridSearchCV

time0 = time()

gamma_range = np.logspace(-10,1,20)
codf0_range = np.linspace(0,5,10)

para_grid = dict(gamma=gamma_range
                , coef0=codf0_range
                )

cv = StratifiedShuffleSplit(n_splits=5, test_size=0.3, random_state=420)
grid = GridSearchCV(SVC(kernel='poly',degree=1,cache_size=5000)
                   , param_grid= para_grid
                    , cv=cv
                   )
grid.fit(X,y)

print("the best para is %s, and the best score is %s"%(grid.best_params_, grid.best_score_))
print(datetime.datetime.fromtimestamp(time()-time0).strftime("%M:%S:%f"))
```

若多项式效果没有rbf/linear好，一般不采用poly

### 软间隔 
>defination: 硬间隔和软间隔
当两组数据是完全线性可分，存在一个决策边界是训练集上的分类误差为0，这两种数据被称为存在“硬间隔”。当两组数据几乎是完全线性可分的，但决策边界在训练集上存在较小的训练误差，这种被称为“软间隔”

最大margin和被分错的样本数量的平衡->松弛系数$\theta$

进一步松弛系数造成的被分错的样本数量增大，应该加上惩罚项来对松弛系数进行限制。

后续一系列的拉格朗日函数、拉格郎日对偶问题实质没有改变，只是拉格朗日乘子$\alpha \leq C$, C为控制惩罚力度的系数。

系数C：0开始，1为比较好；噪声较多时，减少C；有方向性的

实际使用时与核函数的参数gamma,degree等联合调参；

```python
# 对线性核函数
C_range = np.linspace(0.01,5,50)
score = []

for i in C_range:
    clf = SVC(kernel='linear',gamma='auto',C=i, cache_size=5000).fit(Xtrain,Ytrain)
    score.append(clf.score(Xtest,Ytest))
print(C_range[score.index(max(score))],max(score))
plt.plot(C_range, score)
plt.show()
```

```python
# rbf
C_range = np.linspace(0.01,2,50)
score = []

for i in C_range:
    clf = SVC(kernel='rbf',gamma='auto',C=i, cache_size=5000).fit(Xtrain,Ytrain)
    score.append(clf.score(Xtest,Ytest))
print(C_range[score.index(max(score))],max(score))
plt.plot(C_range, score)
plt.show()
```

**进一步了解参数C**：
软间隔让决定两条虚线超平面的支持向量可能时来自同一个类别的样本点， 而硬间隔的时候两条虚线超平面必须由不同类别的支持向量决定。

由于软间隔，选取支持向量的可能选择大大增加了。

C是SVM的重要参数
