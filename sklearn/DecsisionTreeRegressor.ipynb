```python
# 回归树模拟正弦曲线
import matplotlib.pyplot as plt
import numpy as np
```

```python
rng = np.random.RandomState() #  随机数种子
X = np.sort(10*rng.rand(80,1), axis = 0)# x为数组而不是向量80x1
y = np.sin(X).ravel() # y为向量而不是数组 80
y[::8] += 3 * (0.5 - rng.rand(10))
plt.figure()
plt.scatter(X,y)
```

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X,y, test_size = 0.3)
```

```python
regressor1 = tree.DecisionTreeRegressor(random_state = 10, splitter= 'random', max_depth = 2)
regressor2 = tree.DecisionTreeRegressor( max_depth = 5)
regressor1 = regressor1.fit(Xtrain, Ytrain)
regressor2 = regressor2.fit(Xtrain, Ytrain)
score1 = cross_val_score(regressor1, X, y, cv=10)
score2 = cross_val_score(regressor2, X, y, cv=10)
y1_hat = regressor1.predict(X)
y2_hat = regressor2.predict(X)
plt.figure()
plt.scatter(X,y)
plt.plot(X, y1_hat, color = 'red', label = 'max_depth = 2')
plt.plot(X, y2_hat, color = 'green', label = 'max_depth = 5')
plt.legend()
plt.show()
```

```python
X_test = np.arange(0.0, 5.0,0.01)[:, np.newaxis]
X_test
```

```python
y1_hat = regressor1.predict(X_test)
y2_hat = regressor2.predict(X_test)
plt.figure()
plt.plot(X_test, np.sin(X_test), color='blue', label='truth ground')
plt.plot(X_test, y1_hat, color = 'red', label = 'max_depth = 2')
plt.plot(X_test, y2_hat, color = 'green', label = 'max_depth = 5')
plt.legend()
plt.show()
```
