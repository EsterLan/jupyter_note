## NumPy 
>笔记：Python的面向数组计算可以追溯到1995年，Jim Hugunin创建了Numeric库。接下来的10年，许多科学编程社区纷纷开始使用Python的数组编程，但是进入21世纪，库的生态系统变得碎片化了。2005年，Travis Oliphant从Numeric和Numarray项目整合出了NumPy项目，进而所有社区都集合到了这个框架下。

### *What numpy do*

   mostly:
   1. 用于数据整理和清理、子集构造和过滤、转换等快速的矢量化数组运算。
   2. 常用的数组算法，如排序、唯一化、集合运算等。
   3. 高效的描述统计和数据聚合/摘要运算。
   4. 用于异构数据集的合并/连接运算的数据对齐和关系型数据运算。
   5. 将条件逻辑表述为数组表达式（而不是带有if-elif-else分支的循环）。
   6. 数据的分组运算（聚合、转换、函数应用等）。。。。
   7. 广播等高级功能
   


### *Why numpy*

高效处理大数组的数据。
1. NumPy是在一个连续的内存块中存储数据，独立于其他Python内置对象。NumPy的C API可以操作内存，NumPy数组使用的内存更小；
2. 可以在整个数组上执行复杂的计算，不需要Python的for循环处理


### ndarray: N-dimension array object
大数据容器，对整块数据进行数学运算，语法与标量运算一致；同构数据多维容器（数据类型一致dtype&shape）； as called NumPy数组、数组、ndarray




````python
import numpy as np

data = np.random.randn(2,3)
print(data) 
print(data * 10)
print(data + data)
print(data.shape)
print(data.dtype)
````

- ndarray的数据类型： dtype

dtype(数据类型)是一个特殊**对象**，含有ndarray将内存解释为特定数据类型的信息；

dtype是NumPy灵活**交互其它系统**的源泉之一。多数情况下，它们直接映射到相应的机器表示，这使得“读写磁盘上的二进制数据流”以及“集成低级语言代码（如C、Fortran）”等工作变得更加简单

记不住这些NumPy的dtype也没关系，新手更是如此。通常只需要知道你所处理的数据的大致类型是浮点数、复数、整数、布尔值、字符串，还是普通的Python对象即可。当你**需要控制数据在内存和磁盘中的存储方式**时（尤其是对大数据集），那就得了解如何控制存储类型。

- ndarray的运算

1.ndarray可对数据进行批量的元素级处理，NumPy用户称为矢量化

2.切片和索引：
    
    与列表的操作大致相同；高维数组操作时，以轴为参考进行切片及索引:表示选取全轴；
    区别在于切片返回为数据的视图，对切片的修改会扩散到整个选区（广播），对源数据会有影响。why?由于NumPy的设计目的是处理大数据，所以你可以想象一下，假如NumPy坚持要将数据复制来复制去的话会产生何等的性能和内存问题。
    note: 如果你想要得到的是ndarray切片的一份副本而非视图，就需要明确地进行复制操作，例如arr[5:8].copy()。


````python
data1 = np.random.randn(3,4)
arr1 = np.array(data1)
print('Before slice change:arr1=\n', arr1)
# 选取前两列
arr1_slice = arr1[:,:2] 
# 对slice全体值进行改变时加:
arr1_slice[:] = 1
print('arr1_slice=\n',arr1_slice)
print('After slice change: arr1=\n', arr1)
# 选取某个值
print('arr1[1,2]=', arr1[1, 2])
````

注意，在上面所有这些选取数组子集的例子中，返回的数组都是视图。

3.布尔型索引

    总会有需求，考虑以字符为索引对数组进行引用，通过返回布尔型数组对数组进行应用
    note: and和or在布尔型索引中采用 & |;
          如下对二维数据的操作用pandas更为简便
  

````python
names = np.array(['Bob', 'Joe', 'Will', 'Bob', 'Will', 'Joe', 'Joe'])
# 注意维度应该对应
arr = np.array(np.random.randn(7,4))

cond = names == 'Bob'
cond1 = (names== 'Bob') | (names == 'Will')
cond2 = names == 'Will'
print('the orginal array= \n',arr)
print('Bool index array= ', cond)
# 选取‘Bob’行
print('row-Bob= \n',arr[cond])
# ~对cond处理
print('～cond row Bob = \n', arr[~cond])
print('Bob and Will bool index= ', cond1)
# &|对布尔型索引处理，而不是python中的and和or
print('Bob and Will :\n', arr[cond1])
# Test for saperate the conditions
print('test for arr[cond1] == arr[cond | cond2]:\n ', arr[cond1] == arr[cond | cond2 ])
# Bob For entry
print('Bob for entry:取后两列\n',arr[cond,2:])
````

````python
# 通过布尔型数组设值也是常用手段
# 下例将data中的负值设为0
data = np.random.randn(7,4)
print('the original data = \n', data)
data[data<0] = 0
print('set-0 data = \n', data)
````

5.数组转置和轴对换
1.转置是轴对换的轴为2的情况，返回的是源数据的视图（无复制操作）。除了transpose函数外还有T属性
2.对高维数组的transpose需要轴编号的tuple来进行转置；对两个轴进行交换采用ndarray的swapaxes.

````python
# 二维
arr = np.random.randn(6,3)
print('arr=\n',arr)
print('arrT=\n', arr.T)
# 常见操作：内积
print('arr.T*arr= \n', np.dot(arr.T, arr))
````

````python
# 高维
arr1 = np.arange(16).reshape((2,2,4))
print('the original array=\n', arr1)
# 0、1轴交换位置，2轴不变
print('swap 0 1 through transpose:\n', arr1.transpose((1,0,2)))
print('swap 1 2 through swapaxes:\n', arr1.swapaxes(1,2))
````

- 利用数组进行数据处理

   1.Numpy数组将多种数据处理任务表述为简洁的数学表达式；用数组表达式代替编写循环的做法，称为矢量化

````python
points = np.arange(-5,5,0.01)
# np.meshgrid 接收两个一维数组生成两个二维矩阵（对应于两个数组中所有(x,y)）
xs, ys = np.meshgrid(points,points)
ys
````

````python
# 计算sqrt(x^2+y^2)
z = np.sqrt(xs ** 2 + ys ** 2)
# sqrt(-5^2+-5^2)sqrt(-4.99^2+-5^2)对应为上平方开根
z
````

````python
import matplotlib.pyplot as plt
plt.imshow(z, cmap = plt.cm.gray);plt.colorbar()
plt.title("Image plot of $\sqrt{x^2 + y^2}$ for a grid of values")
````

2.条件逻辑表述为数组运算：numpy.where

    numpy.where函数是三元表达式x if condition else y的矢量化版本。
```python
result = [x if c else y for x,y,c in zip(xarr, yarr, cond)]
```
equal to
```python
result = np.where(cond, xarr, yarr)
```
    另:np.where的第二个和第三个参数数组大小可不等，也可以是标量；    

````python
arr = np.random.randn(4,4)
print('the original arr= \n', arr)
# arr中正值用1代替
np.where(arr>0, 1, arr)
````

3.数学统计方法

    对一个轴上或整个数组的数据进行处理，如sum、mean、std等聚合计算AKA aggregation or reduction，当成实例方法调用/np的顶级函数；
    还可接受轴数作为参数，返回少一维的数组；
    累加函数cumsum&累乘函数cunprod返回同样大小的数组
    note: sum mean std/var(自由度可调) min/max argmin/argmax(最大最小元素的索引) cumsum/cumprod

````python
import numpy as np
arr = np.random.randn(7,4)
print('arr = \n',arr)
# 调用方式
print('mean of arr:', np.mean(arr))
print('sum of colums of arr:', arr.sum(axis = 0))
# cumsum返回同样大小数组
print('累加\n', arr.cumsum())
````

4.方法for布尔型数组

    上述方法中bool型被转换为1/0（True/False），sum可对其求和以统计true的个数；
    另外，有常用于布尔型数组的函数any(是否有true)，all(是否全为True)
    note: any/all也可被用于非布尔型，此时非零元素为true

````python
# 统计True的个数
arr = np.random.randn(100)
print('#True:', (arr>0).sum())

# any all
arr_b = np.array([False, True, False, False])
print('any True in array?', np.any(arr_b))
print('all True array?', np.all(arr_b))
````

- linera algebra

    点乘dot,x.dot(y)相当于np.dot(x,y);numpy的linalg提供标准的矩阵分解、求逆、行列式等。
    note: np.inv/qr不存在即numpy和numpy.linalg不是一个东西
          diag(返回对角元素/一维转换为方阵) dot trace det eig inv pinv(Moore-Penrose逆) qr（QR分解）svd solve(Ax=b,A为方阵) lstsq(Ax=b的最小二乘解)

````python
from numpy.linalg import qr, inv
X = np.random.randn(5,5)
mat = np.dot(X.T,X)
# inv
print('inverse of X^TX = \n',inv(mat))
# qr
q,r = qr(mat)
print('upper traiangular matrix= \n', r)
````
